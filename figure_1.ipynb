# Packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Figures

## settings

```python
plt.rcParams.update({"font.family": "Arial", "font.size": 8, 'axes.titlesize': 'medium',})
```

## colors

```python
celltype_colors = {
    'DETC': '#1f78b4',
    'LC': '#e31a1c'
}
```

## lineplot DETC and LC quantification

```python
df = pd.read_excel('/path/to/data/Bl6_Quantification.xlsx', header=0)
```

```python
df = df.melt(
    id_vars=['timepoint', 'replicate'],
    var_name='cell_type',
    value_name='count'
)

df['cell_type'] = df['cell_type'].str.removesuffix('s')
```

```python
conversion_in_mm = 25.4
width = 80
height = 60

fig, ax = plt.subplots(1,1, figsize=(80/conversion_in_mm, 60/conversion_in_mm), dpi=300, layout = 'constrained')

sns.lineplot(
    df,
    x = 'timepoint',
    y = 'count',
    hue = 'cell_type',
    palette = celltype_colors,
    ax=ax
)

sns.stripplot(
    df,
    x = 'timepoint',
    y = 'count',
    hue = 'cell_type',
    legend=None,
    palette = celltype_colors,
)

ax.legend(title='Cell type', frameon=False)
ax.set(
    xlabel = '',
    ylabel = 'Cells/mm$$',
    title = 'Quantification'
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

## lineplot KI67 quantification

```python
df_ki67 = pd.read_excel('/path/to/data/Ki67_quantification_tidy.xlsx', header=0)
```

```python
conversion_in_mm = 25.4
width = 80
height = 60

fig, ax = plt.subplots(1,1, figsize=(80/conversion_in_mm, 60/conversion_in_mm), dpi=300, )

sns.lineplot(
    df_ki67,
    x = 'Timepoint',
    y = 'Ki67+',
    hue = 'Celltype',
    palette = celltype_colors,
    ax=ax
)

sns.stripplot(
    df_ki67,
    x = 'Timepoint',
    y = 'Ki67+',
    hue = 'Celltype',
    legend=None,
    palette = celltype_colors,
)

ax.legend(title='Cell type', frameon=False)
ax.set(
    xlabel = '',
    ylabel = 'Ki67+ [%]',
    title = 'Quantification Ki67+'
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```
